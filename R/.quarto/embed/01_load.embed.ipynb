## Data load

Loading libraries:

```python
#| message: false

library(tidyverse)
library(here)
```

Loading the data from URL into R:

```python
#| label: "Load_data"

data_url <- "https://archive.ics.uci.edu/static/public/565/bone+marrow+transplant+children.zip"

#Creating folder for data if not already done
raw_dir <- here("data/_raw/")
if( !dir.exists(raw_dir) ){
  dir.create(path = "data")
  dir.create(path = raw_dir)
}

#Reading data file, if it has not been done before
if( !file.exists(str_c(raw_dir, "bone-marrow.arff")) ){
  download.file(
      url = data_url,
      destfile = str_c(raw_dir,"bone-marrow.arff")
      )
}
```
